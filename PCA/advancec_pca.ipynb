```python
import pandas as pd
pd.set_option('max_rows', 500)
pd.set_option('max_columns', 500)
df = pd.read_csv('sugarcane_spatial_cluster.csv')
x1 = df[['incidencia_RSD','longitude', 'latitude', 'area','edad',
       'incidencia_LSD', 
       'NDVI', 'RH2M', 'PRECTOTCORR', 'WS2M', 'PAR', 'T2M',
       'T2M_MAX', 'T2M_MIN', 'densidad_aparente_0-30cm_mean',
       'intercambio_catiónico_0-30cm_mean', 'fraccion_gruesa_0-30cm_mean',
       'arcillas_0-30cm_mean', 'nitrogeno_0-30cm_mean', 'ph_0-30cm_mean',
       'arena_0-30cm_mean', 'limo_0-30cm_mean', 'carbono_organico_0-30cm_mean', 'cluster']]
```

```python
df_drop=x1.drop(labels=['incidencia_RSD', 'incidencia_LSD'],axis=1)
cols = df_drop.columns.tolist()
cols.insert(0, cols.pop(cols.index('cluster')))
df_drop = df_drop.reindex(columns= cols)
df_drop
```

```python
df_drop=df_drop.drop(labels=['cluster','longitude', 'latitude'],axis=1)
```

```python
df_drop
```

```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from advanced_pca import CustomPCA

X_std = StandardScaler().fit_transform(df_drop)
```

## The significant feature mask in custom_pca.py is > 5

```python
# fit pca objects with option selecting only significant features
significant_pca5 = (CustomPCA(n_components=8, feature_selection='significant')
                    .fit(X_std))

# print selected features based on weights and communalities significance
print('Communalities:\n', significant_pca5.communalities_)
print('\nSelected Features:\n',
      np.asarray(df_drop.columns)[significant_pca5.get_support()])

# execute dimensionality reduction and pring dataset shapes
print('\nOriginal dataset shape:', X_std.shape)
print('Reduced dataset shape:', significant_pca5.transform(X_std).shape)
```
